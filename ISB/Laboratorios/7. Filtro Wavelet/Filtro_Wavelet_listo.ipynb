```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import firwin, lfilter
from tqdm import tqdm
from scipy import signal
import pywt
```

# Para el ECG

ECG en reposo:

```python
# leyendo archivo texto
archivo_txt = "/content/NormalBreath30s.txt"
datos_ecg = np.loadtxt(archivo_txt)
senal = datos_ecg[:, 5]  # Sexta columna

# arreglo para los tiempos
frecuencia_muestreo = 1000 #Hz
tiempo = np.arange(0,np.size(senal))/frecuencia_muestreo

#Asimismo, la funcion de tranferencia
senal=(((senal/1024)-0.5)*3.3)/(1100)
senal=senal*frecuencia_muestreo;

# Aplicar la transformada wavelet
coeffs = pywt.wavedec(senal, 'db6', level=6)

# Visualizar todos los niveles de la wavelet
plt.figure(figsize=(8, 12))
for i, coeff in enumerate(coeffs):
    plt.subplot(len(coeffs) + 1, 1, i + 1)
    plt.plot(coeff,'m')
    plt.ylabel(f'Coeff (N={i})')

plt.subplots_adjust(hspace=0.5)

# Filtrar los coeficientes
sigma = np.median(np.abs(coeffs[-1])) / 0.6745  # Estimaci칩n robusta del ruido

# Calcular el umbral universal
N = len(senal)
threshold = sigma * np.sqrt(2 * np.log(N))
filtered_coeffs = [pywt.threshold(c, threshold, mode='soft') for c in coeffs]

# Reconstruir la se침al filtrada
filtered_signal = pywt.waverec(filtered_coeffs, 'db6')

plt.suptitle('Niveles (N) del 0-8 respectivamente', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.97])

plt.show()

plt.figure(figsize=(8, 12))
plt.subplot(2, 1, 1)
plt.plot(tiempo[20000:24000], senal[20000:24000],'m')
plt.title('Se침al original ECG Reposo')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')

plt.subplot(2, 1, 2)
plt.plot(tiempo[20000:24000], filtered_signal[20000:24000],'m')
plt.title('Se침al filtrada ECG Reposo')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
```

# PARA EMG:

EMG en reposo:

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Cargar la se침al EMG
arrayemg = np.genfromtxt("text_EMG.txt", delimiter="\t", skip_header=3, missing_values=0)
Fs = 1000  # Frecuencia de muestreo en Hz

# Seleccionar los primeros 10 segundos de la se침al
duration = 10 # Duraci칩n en segundos
samples = Fs * duration # N칰mero de muestras correspondientes a los primeros 10 segundos
signalemg = arrayemg[:samples, 5]
n = len(signalemg)
t = np.arange(0, n / Fs, 1 / Fs)

# An치lisis de Wavelets
wavelet = 'db4'
levels = 5
coeffs = pywt.wavedec(signalemg, wavelet, level=levels)

# Visualizaci칩n
plt.figure(figsize=(15, 20))

# Se침al EMG Original
plt.subplot(5, 1, 1)
plt.plot(t, signalemg, label='Original EMG')
plt.title('Se침al EMG Original')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()

# Coeficientes de Wavelet Detalle (Niveles 1-4)
for i in range(1, 5):
    plt.subplot(5, 1, i + 1)
    plt.plot(coeffs[i], label=f'Detalle Nivel {i}')
    plt.title(f'Coeficientes de Detalle Nivel {i}')
    plt.xlabel('Muestras')
    plt.ylabel('Amplitud')
    plt.legend()

plt.tight_layout()
plt.show()
```
